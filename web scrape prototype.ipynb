```python
from requests import get
from bs4 import BeautifulSoup
from collections import OrderedDict
import time
import random
import pandas as pd
import re
from IPython.core.display import clear_output
```

```python
start_time = time.time()
pages = 5


anime_name = []
username = []
review_date = [] 
episodes_seen = []
episodes_total = []
overall_rating = []
story_rating = []
animation_rating = []
sound_rating = []
character_rating = []
helpful_counts = []
review_body = []
enjoyment_rating = []

for j in range(pages):
    
    #Makes the request
    url = 'https://myanimelist.net/reviews.php?t=anime&p={}'.format(j+1)
    print(url)
    
    response = get(url)
    if response.status_code != 200:
        warn('Request: {}; Status code: {}'.format(requests, response.status_code))
        
    #Creates the soup object    
    html_soup = BeautifulSoup(response.text, 'html.parser')
    review_containers = html_soup.find_all('div', class_ = 'borderDark')
    
    #Loops through the containers on a page
    for container in review_containers:
        review_element = container.div
        
        anime_name.append(review_element.find('a', class_='hoverinfo_trigger').text)         
        username.append(review_element.find_all('td')[1].a.text)
        review_date.append(review_element.div.div.text)
        episodes_seen.append(review_element.div.find_all('div')[1].text.strip().split(' ')[0])
        episodes_total.append(review_element.div.find_all('div')[1].text.strip().split(' ')[2])
        overall_rating.append(review_element.div.find_all('div')[2].text.strip().split('\n')[1])
        story_rating.append(review_containers[1].find_all('td', class_='borderClass')[3].text)
        animation_rating.append(review_containers[1].find_all('td', class_='borderClass')[5].text)
        sound_rating.append(review_containers[1].find_all('td', class_='borderClass')[7].text)
        character_rating.append(review_containers[1].find_all('td', class_='borderClass')[9].text)
        
        enjoyment_value = review_containers[1].find_all('td', class_='borderClass')[11].text
        enjoyment_rating.append(enjoyment_value)
        helpful_counts.append(review_element.find('span').text)
        
        body = container.select('div.spaceit.textReadability.word-break.pt8')[0].text.strip().replace('\n', ' ')
        cleaned_body = re.sub(r'more pics      Overall .*   Enjoyment {}'.format(enjoyment_value), '', body)
        review_body.append(cleaned_body)
        
    #Provide stats for monitoring
    current_time = time.time()
    elapsed_time = current_time - start_time
    requests = j+1    
        
    print('Request Completed: {}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))    
    if requests == pages:
        break
    print('Loop Pausing...')    
    time.sleep(random.uniform(20, 30))    
    clear_output(wait = True)
    
    
    
anime_ratings = pd.DataFrame(OrderedDict((
                              ('username', username),
                              ('anime_name', anime_name),
                              ('review_date', review_date),
                              ('episodes_seen', episodes_seen),
                              ('episodes_total', episodes_total),
                              ('overall_rating', overall_rating),
                              ('story_rating', story_rating),
                              ('animation_rating', animation_rating),
                              ('sound_rating', sound_rating),
                              ('character_rating', character_rating),
                              ('enjoyment_rating', enjoyment_rating),    
                              ('helpful_counts', helpful_counts),
                              ('review_body', review_body))
                            )
                )                             
anime_ratings.to_csv('anime_ratings.csv', index=False)   
```
